# dimtensor Quick Start

Welcome to dimtensor! This notebook demonstrates basic usage.

```python
from dimtensor import DimArray, units
import numpy as np

# Create unit-aware arrays
distance = DimArray([1, 2, 3], units.m)
time = DimArray([1, 2, 3], units.s)

# Dimensional analysis
velocity = distance / time
print(f"Velocity: {velocity}")
print(f"Unit: {velocity.unit}")
```

```python
# Physical constants
from dimtensor.constants import c, h, k_B

print(f"Speed of light: {c}")
print(f"Planck constant: {h}")
print(f"Boltzmann constant: {k_B}")
```

```python
# PyTorch integration
try:
    from dimtensor.torch import DimTensor
    import torch
    
    t = DimTensor([1.0, 2.0, 3.0], units.m, requires_grad=True)
    print(f"PyTorch DimTensor: {t}")
    print(f"CUDA available: {torch.cuda.is_available()}")
except ImportError:
    print("PyTorch not available in this image")
```
